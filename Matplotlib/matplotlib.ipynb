# Horizontal Subplots

```python
plt.subplot(1,2,1)
plt.plot([1, 2, 3, 4], [1,4,9,16], "go")
plt.title("First SubPlot")
plt.xlabel("X Value")
plt.ylabel("Y Value")

plt.subplot(1,2,2)
x=np.arange(1,5)
y=x**3
plt.plot(x, y, 'r^')
plt.title("Second SubPlot")
plt.xlabel("X Value")
plt.ylabel("Y Value")

plt.show()
```

# Vertical Subplots

```python
plt.subplot(2,1,1)
plt.plot([1, 2, 3, 4], [1,4,9,16], "go")
plt.title("First SubPlot")
plt.xlabel("X Value")
plt.ylabel("Y Value")

plt.subplot(2,1,2)
x=np.arange(1,5)
y=x**3
plt.plot(x, y, 'r^')
plt.title("Second SubPlot")
plt.xlabel("X Value")
plt.ylabel("Y Value")

plt.show()
```

# Manage Subplots 

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(6,6))
axes[0].plot([1, 2, 3, 4], [1,4,9,16], "go")
axes[0].set_title("First SubPlot")
axes[0].set_xlabel("X Value")
axes[0].set_ylabel("Y Value")

x=np.arange(1,5)
y=x**3
axes[1].plot(x, y, 'r^')
axes[1].set_title("Second SubPlot")
axes[1].set_xlabel("X Value")
axes[1].set_ylabel("Y Value")

fig.tight_layout() # Or equivalently,  "plt.tight_layout()"

plt.show()
```

# Vertical Plot

```python
product=["A", "B", "C"]
sell_average=[20, 50, 30]

plt.bar(product, sell_average, color="green")
plt.title("Product Sell Average")
plt.xlabel("Product")
plt.ylabel("Sell Average")
```

# Horizontal Bar Plot

```python
product=["A", "B", "C"]
sell_average=[20, 50, 30]
sell_variance=[5, 2, 3]

plt.barh(product, sell_average, xerr=sell_variance, color="green")
plt.title("Product Rotability")
plt.xlabel("Product Class")
plt.ylabel("Value")
```

# Histogram Plot

```python
x= np.random.randn(1000)

plt.title("Histogram Plot")
plt.xlabel("Random data")
plt.ylabel("Frecuency")
plt.hist(x, 10)

plt.show()
```

# Scatter Plot

```python
height = [1,5,9,10, 12]
weight = [20,80,50,10,65]

plt.scatter(height, weight)
plt.title("Scatter Plot")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

# 3D Plots

```python
ax = plt.axes(projection='3d')

ax.scatter3D(height, weight)
ax.set_xlabel("Height")
ax.set_ylabel("Height")
plt.show()
```

# Plotting with keyword strings

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

# Working with multiple figures and axes

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

# Working with text

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

# Annotating text

```python
ax = plt.subplot()

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )

plt.ylim(-2, 2)
plt.show()
```
